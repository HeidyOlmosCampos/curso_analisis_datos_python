### **4. BUCLES**

```python
############################# BUCLES FOR IN #####################################
#Sintaxis:
#For elemento in elementos:
#   Acciones

iterador = list(range(0, 11))
print(iterador)
suma = 0

for numero in iterador:
    suma = suma + numero

print(suma)

lista_palabras : list = ['Instagram', 'tiktok', 'youtube']

for palabra in lista_palabras:
    print(palabra)

############################# BUCLES WHILE #####################################
#Sintaxis:
#While condicion:
#   Acciones

suma = 100
while suma > 0:
    print(suma)
    suma = suma - 10

############################# SENTENCIAS BREAK, PASS Y CONTINUE #####################################

#Break: Detiene la ejecución del bucle.

while True:
    break

suma = 100
while suma > 0:
    if suma == 90:
        print("La var suma es 90")
        break
    print(suma)
    suma = suma - 1


#Pass: Se utiliza para dejar vacío un trozo de código que todavía no queremos especificar.

for numero in iterador:
    pass

print("Hola mundo")

#Continue: Pasa a la siguiente iteración del bucle.

for numero in iterador:
    if numero % 2 == 0:
        continue
    print(numero)

```

### **5. FUNCIONES**

```python
# Sintaxis: def nombre_funcion(parametro1 : Tipo1, parametro2: Tipo2) -> TipoDatoRetorno:
#               Acciones
#               return Valor (Opcional)

############################# FUNCIÓN SIN PARÁMETROS #####################################
def funcion_basica() -> None:
    print('Hola mundo')

funcion_basica()


############################# FUNCIÓN CON PARÁMETROS #####################################

# Python es un lenguaje con tipado dinámico, es decir, no hace falta indicar que tipo de datos va a recibir o devolver una función ya que Python
# va a inferir los tipo de datos en tiempo de ejecución. Es muy recomendable indicar los tipos de datos para aumentar la legibilidad del código.
# Es más, aunque tu le indiques el tipo de parámetros y luego le pasas otro tipo de datos NO va a dar un fallo.

def suma(numero1: int, numero2: int) -> int:
    return numero1 + numero2

print(suma(5, 6))


############################# FUNCIÓN CON PARÁMETROS CON VALORES POR DEFECTO #####################################

def sacar_dinero(dinero_actual: int, dinero_a_retirar: int = 50) -> int:
    if dinero_actual >= dinero_a_retirar:
        dinero_actual = dinero_actual - dinero_a_retirar
    else:
        print("No tienes suficiente saldo en la cuenta.")
    return dinero_actual

saldo_cuenta: int = 120
saldo_actual: int = sacar_dinero(saldo_cuenta, 10)
print(saldo_actual)

############################# FUNCIÓNES QUE DEVUELVE MÁS DE UN VALOR #####################################
def dos_elementos(lista:list) -> str:
    if len(lista) >= 2:
        return lista[0], lista[1]
    return "", ""

nombres : list = ['heidy', 'Juan', 'Leidy']

nombre1, nombre2 = dos_elementos(nombres)
print(nombre1)
print(nombre2)
```

### **6. PROGRAMACIÓN ORIENTADA A OBJETOS**

```python

# Sintaxis:

# class NombreClase():
#   Atributos
#   def __init__(self):
#       Inicializar atributos (constructor)
#   Setters and getters
#   Metodos de clase
#

# Implementación de la clase mascota para un videojuego de animales.

class Mascota():  
    
    ############################# ATRIBUTOS DE LA CLASE #####################################
    __nombre : str          #private
    _puntos_salud : int     #protected
    __nivel : int
    __edad : int

    ############################# CONSTRUCTOR DE LA CLASE #####################################
    def __init__(self, nombre : str, puntos_salud : int, edad : int):
        self.__nombre = nombre
        self._puntos_salud = puntos_salud
        self.__nivel = 1
        self.__edad = edad


    ############################# METODOS GETTERS AND SETTERS #####################################

    # Los metodos getters y setters sirven para ofrecer la funcionalidad de obtener el valor o cambiar el valor de los atributos que queremos darle 
    # la posibilidad al programador que utilice la clase.

    # Sintaxis getters: (Los atributos deben ser privados o protegidos, sino no tendría sentido hacer un get o set ya que podemos modificar y
    # consultar los atributos directamente, es más daría un bucle infinito.)
    
    # def get_atributo(self) -> tipo_atributo:
    #   return self.__nombre_atributo

    # Sintaxis setters:
    # def set_atributo(self, value) -> None:
    #   self.__nombre_atributo = value

    def get_nombre(self) -> str:
        return self.__nombre
    
    def set_nombre(self, nombre : str):
        self.__nombre = nombre

    def get_puntos_salud(self) -> int:
        return self._puntos_salud
    
    def set_puntos_salud(self, puntos_salud : int):
        self._puntos_salud = puntos_salud

    def get_edad(self) -> int:
        return self.__edad
    
    def get_nivel(self) -> int:
        return self.__nivel

  ############################# METODOS DE CLASE #####################################

  # Método que sirve para aumentar la salud a tu mascota.

    def alimentar_mascota(self, comida : str) -> None:
        info_comida : dict = {'comida_buena': 10, 'comida_genial': 20, 'comida_excelente': 30}

        puntos_salud_comida = info_comida.get(comida, 0)
        self._puntos_salud = self._puntos_salud + puntos_salud_comida

        print(f"Mmm.... que rico, gracias por alimentarme, ahora tengo {self._puntos_salud} puntos de salud")

  # Método que sirve para aumentar de nivel a tu mascota en función del torneo que haya ganado.
    def ganar_torneo(self, tipo_torneo: str) -> None:
        info_torneo : dict = {'regional': 1, 'nacional': 2, 'internacional': 3}

        niveles_torneo = info_torneo.get(tipo_torneo, 0)
        self.__nivel = self.__nivel + niveles_torneo

        print(f"Enhorabuena {self.__nombre} has ganado un torneo {tipo_torneo} y ahora estas en el nivel {self.__nivel}")

mascota_1 : Mascota = Mascota('Nemo', 100, 2)

print(mascota_1.get_nombre())
mascota_1.alimentar_mascota('comida_buena')
mascota_1.ganar_torneo('internacional')
    
```
